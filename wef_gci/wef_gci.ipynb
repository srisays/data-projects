# World Economic Forum's Global Competitiveness Index

Launched in 1979 by Klaus Schwab, Founder and Executive Chairman of the World Economic Forum, and initially covering 16 countries, the Global Competitiveness Report 2014-2015, today assesses the competitiveness landscape of 144 economies, providing insight into the drivers of their productivity and prosperity. Since 2005, the World Economic Forum has based its competitiveness analysis on the Global Competitiveness Index (GCI), a comprehensive framework that measures the microeconomic and macroeconomic foundations of national competitiveness, grouped into 12 categories.

The data is available for downlad on the [World Economic Forum website](http://reports.weforum.org/global-competitiveness-report-2014-2015/) in XLS format. The data has been cleaned and converted to Postgres SQL format. The PostgreSQL database dump is [available on Github](https://github.com/srisays/data-projects/tree/master/wef_gci).

```python
import pandas
import vincent
import sqlalchemy
import IPython.display
from vincent import AxisProperties, PropertySet, ValueRef
from IPython.html import widgets 
from IPython.html.widgets import interact
vincent.core.initialize_notebook()
engine = sqlalchemy.create_engine('postgresql://sri@localhost:5432/wef_gci')
```

## Overall GCI Ranking

Based on the GCI Framework the 12 Pillars of Competitiveness are Institutions, Infrastructure, Macroeconomic environment, Health & primary education, Higher education & training, Goods market efficiency, Labor market efficiency, Financial market development, Technological readiness, Market size, Business sophistication and Innovation. The top 15 country rankings based on all the 12 pillars are as below. 

```python
pandas.io.sql.read_sql(
"""SELECT rank, edition, entity_name 
FROM gci 
JOIN entities on entities.entity_id = gci.country_code
WHERE placement_id = 158 and edition = 2014
ORDER BY rank ASC
Limit 15;""",
con=engine)
```

## Country Rankings by 12 Pillars 

Let's analyze the rankings further by investigating how each country stacks up against the other based on the individual  pillars - Institutions, Infrastructure, Macroeconomic environment, Health & primary education, Higher education & training, Goods market efficiency, Labor market efficiency, Financial market development, Technological readiness, Market size, Business sophistication and Innovation.

```python
def rankings_by_pillar_id(pillar_id):
    x = pandas.io.sql.read_sql(
        sql="""SELECT edition, entity_name, rank FROM gci
        JOIN entities on entities.entity_id = gci.country_code
        WHERE placement_id = %(pl_id)s and edition = 2014
        ORDER BY rank ASC
        Limit 15;""",
        params={'pl_id': pillar_id},
        con=engine)
    print x
    return x

interact(rankings_by_pillar_id, pillar_id=widgets.DropdownWidget(values=
{
'Pillar 1 - Institutions': 31, 
'Pillar 2 - Infrastructure': 43, 
'Pillar 3 - Macroeconomic Environment': 49,
'Pillar 4 - Health & Primary Education': 62,
'Pillar 5 - Higher Education & Training': 75,
'Pillar 6 - Goods Market Efficiency': 96,
'Pillar 7 - Labor Market Efficiency': 109,
'Pillar 8 - Financial Market Development': 120,
'Pillar 9 - Technological Readiness': 130,
'Pillar 10 - Market size': 137,
'Pillar 11 - Business Sophistication': 148,
'Pillar 12 - Innovation': 156
})) 

```

## Ranking by SubIndexes 

The 12 pillars are further grouped in to 3 subindexes. The Basic Requirements SubIndex comprises of the following pillars - Institutions, Infrastructure, Macroeconomic environment, Health & primary education. The Efficiency Enhancers SubIndex contains Higher education & training, Goods market efficiency, Labor market efficiency, Financial market development, Technological readiness and Market size. The ￼Innovation and Sophistication factors SubIndex includes Business sophistication and Innovation pillars. 

```python
def rankings_by_subindex(subindex_id):
    x = pandas.io.sql.read_sql(
        sql="""SELECT edition, entity_name, rank from gci
        JOIN entities ON entities.entity_id = gci.country_code
        WHERE placement_id = %(pl_id)s and edition = 2014
        ORDER BY rank ASC
        limit 15;""",
        params={'pl_id': subindex_id},
        con=engine)
    print x
    return x 

interact(rankings_by_subindex, subindex_id=widgets.DropdownWidget(values=
{
'SubIndex - Basic Requirements': 63,
'SubIndex - Efficiency Enhancers': 138,
'SubIndex - Innovation and Sophistication Factors': 157
})) 
```

## GCI Ranking by BRICS

How do the BRICS rank? The ranking of the five major emerging national economies: Brazil, Russia, India, China, and South Africa based on all the 12 pillars.

```python
pandas.io.sql.read_sql(
"""SELECT rank, edition, entity_name 
FROM gci 
JOIN entities on entities.entity_id = gci.country_code
WHERE placement_id = 158 and edition = 2014 and gci.country_code IN ('BRA','RUS','IND','CHN','ZAF')
ORDER BY rank ASC;""",
con=engine)
```

## India Vs Asia

How does India fare against the rest of Asia?

```python
india_vs_asia = pandas.io.sql.read_sql(
"""SELECT ind.pillar as pillar, 
    round(eda_average::decimal, 1) as "Emerging and Developing Asia", 
    round(india_value::decimal, 1) as "India" FROM 
    (
        SELECT series_name as pillar, 
                value as india_value
        FROM gci 
        JOIN placements ON placements.placement_id = gci.placement_id
        JOIN entities ON entities.entity_id = gci.country_code
        WHERE 
            placements.placement_id IN (31,43,49,62,75,96,109,120,130,137,148,156)
            AND edition = 2014 
            AND gci.country_code ='IND'
    ) AS ind 
    JOIN
    (
        SELECT series_name as pillar, AVG(value) as eda_average
        FROM gci 
        JOIN placements ON placements.placement_id = gci.placement_id
        JOIN entities ON entities.entity_id = gci.country_code
        WHERE 
            placements.placement_id IN (31,43,49,62,75,96,109,120,130,137,148,156) 
            AND gci.edition = 2014 
            AND entities.imf_region ='Emerging and Developing Asia'
            GROUP BY pillar
            ORDER BY pillar) AS eda
    ON eda.pillar = ind.pillar;""", 
con=engine)
india_vs_asia
```

```python
india_vs_asia.set_index('pillar', inplace=True)
bar = vincent.GroupedBar(india_vs_asia[['Emerging and Developing Asia', 'India']], width=720, height=360)
#rotate x axis labels
ax = AxisProperties(
         labels = PropertySet(angle=ValueRef(value=90), align=ValueRef(value="left")))
bar.axes[0].properties = ax
bar.legend(title='Region')
bar.display()
```
